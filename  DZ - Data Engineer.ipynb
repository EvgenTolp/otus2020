```python
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option('max_colwidth', 10000) #сделаем ширину столбца и ячейки больше
pd.set_option('max_rows', 800)
```

```python
Category = [
['Количество вакансий ', 674, 39500, 12000, 23478],
['Мин ЗП', 570, 10183, 4200, 7750],
['Макс ЗП', 7142, 11700, 9750, 11833]
]
agg = ['parameters','НН.ru', 'Linkedin', 'Glassdoor', 'Monster']
```

```python
stat = pd.DataFrame(data = Category , columns = agg)
stat
```

# <center> 1. Сравнение зарплат и предложений Data Engineer </center>

```python
ax = (
    stat
    .pivot_table(index='parameters', values= ['НН.ru', 'Linkedin', 'Glassdoor', 'Monster'], aggfunc='sum')
    .sort_values('parameters',ascending =  True).head(5)
    .plot(figsize=(17, 5), kind = 'bar', grid=True, title='Сравнение вакансий и зарплат в срезе агрегаторов работы')
)
```

```python
CategoryHH = [
['ИТ, системная интеграция', 336],
['Финансовый сектор', 105],
['Розничная торговля', 47],
['Телекоммуникации, связь', 36],
['СМИ, маркетинг, реклама, BTL, PR, дизайн, продюсирование', 18],    
]
aggHH = ['parameters','НН.ru']
statHH = pd.DataFrame(data = CategoryHH , columns = aggHH)
```

```python
CategoryHH_gr = (
    statHH
    .pivot_table(index='parameters', values='НН.ru')
    .sort_values('НН.ru',ascending = False)
    .plot(y = 'НН.ru',
          kind = 'pie',  legend=False, autopct='%1.1f%%',  figsize = (7.5,10), startangle=45,
         title='HH.RU - Потребность в Data Engineer по разным отраслям', textprops=dict(size=15),)
)
```

 # <center> 3. Какие инструменты, языки, фреймворки нужны для Data Engineer: </center>

```python
Requirements = [
['SQL', 100],
['Python', 85],
['Experience 3+', 88],
['Amazon AWS, RedShift and other cloud platforms', 90],
['Scala', 45], 
['Kubernetes, Docker, and Git', 70], 
['Building ‘big data’ pipelines, ETL/ELT', 95],
['BI and integrations', 20],
['Hadoop (Hive, Impala, HDFS)', 85], 
['Communication skills', 10],    
]
aggRequirements = ['parameters','Requirements']
statRequirements = pd.DataFrame(data = Requirements, columns = aggRequirements)
```

```python
CategoryRequirements = (
    statRequirements
    .pivot_table(index='parameters', values='Requirements')
    .sort_values('Requirements',ascending = True)
    .plot(y = 'Requirements',
          kind = 'pie',  legend=False, autopct='%1.1f%%',  figsize = (10,10),startangle=57,
         title='Популярные требования к Data Engineers', textprops=dict(size=14),)
)
```
